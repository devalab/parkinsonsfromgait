```python
def updateHeight(x):
    if x>2:
        return x/100
    else: 
        return x
```

```python
files = os.listdir("../dataset/raw/")
walks = []
for i in range(len(files)):
    if len(files[i])==13:
        walks.append(files[i])
print (len(walks))
print (walks[1])
print (walks[1][7:9])
```

```python
walksinfo = []
for i in walks:
    walk = {}
    walk["Walk Name"] = i
    walk["Patient ID"] = i[0:6]
    temp = df[df['ID']==walk["Patient ID"]]
    walk["Gender"] = temp["Gender"].values[0]
    walk["Age"] = temp["Age"].values[0]
    walk["Height"] = temp["Height"].values[0]
    walk["Weight"] = temp["Weight"].values[0]
    walk['HoehnYahr'] = temp['HoehnYahr'].values[0]
    walk['UPDRS'] = temp['UPDRS'].values[0]
    walk["UPDRSM"] = temp["UPDRSM"].values[0]
    walk["TUAG"] = temp["TUAG"].values[0]
    walk["Speed"] = temp["Speed_"+i[7:9]].values[0]
    if i[7:9] == "10":
        walk["Extra Task"] = 1
    else :
        walk["Extra Task"] = 0
    if "Pt" in i:
        walk["Class"] = 1
    else:
        walk["Class"] = 0

#     print (walk)
    walksinfo.append(walk)
#     break
```

```python
walkdf = pd.DataFrame(walksinfo)

walkdf
```

```python
walkdf = walkdf.sort_values("Walk Name")
# walkdf.sort_values("Walk Name")
temp - set(walkdf["Patient ID"].unique())
```
